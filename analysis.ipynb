```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
```

```python
results_dir = Path("results/testmore")
```

```python
def plot_results(results_dir: Path):
    fig, ax = plt.subplots()
    gens = range(1, len(list(results_dir.glob("*.csv")))+1)

    worsts = []
    means = []
    bests = []

    for gen in gens:
        df = pd.read_csv(results_dir / f"{gen}.csv")
        worsts.append(-1 * df["acc"].max())
        means.append(-1 * df["acc"].mean())
        bests.append(-1 * df["acc"].min())

    ax.plot(gens, bests, label="best")
    ax.plot(gens, means, label="mean")
    ax.plot(gens, worsts, label="worst")

    ax.set_title("Accuracy over generations")
    ax.set_xlabel("Generation")
    ax.set_ylabel("Accuracy")
    ax.legend()

    plt.show()

plot_results(results_dir)
```
